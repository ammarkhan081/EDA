```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Load the dataset

```python
file_path = "C:/Users/FCT/Downloads/salary_02.csv"
salary_data = pd.read_csv(file_path)
salary_data
```

# Checking null values

```python
salary_data.isnull().sum()
```

# Data Cleaning

```python
salary_data['remote_ratio'].fillna(value=salary_data['remote_ratio'].median(), inplace=True)
```

```python
salary_data.isnull().sum()
```

# Descriptive Statistics

```python
salary_data['salary'].describe()
```

# Histogram

```python
plt.figure(figsize=(15, 5))
sns.histplot(salary_data['salary'], bins=15, kde=False, color='red')
plt.title("Histogram of Salaries")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

# Box Plot

```python
plt.figure(figsize=(8, 4))
sns.boxplot(y=salary_data['salary'])
plt.title("Box Plot of Salaries")
plt.xlabel("Salary")
plt.show()
```

# Scatter Plot

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(x=salary_data['experience'], y=salary_data['salary'], color='black')
plt.title("Scatter Plot: Experience vs Salary")
plt.show()
```

# PDF using KDE

```python
plt.figure(figsize=(8, 4))
sns.kdeplot(salary_data['salary'], fill=True, color='blue', alpha=0.6)
plt.title("Probability Density Function (PDF) of Salaries")
plt.show()
```

### Conclusion:
1. The salary data is concentrated within a specific range, as shown by both the histogram and PDF.
2. The PDF highlights that the highest density of salaries lies within a particular range (e.g., 50k-75k).
3. The box plot reveals potential outliers, indicating some employees earn significantly more or less than the majority.
4. The scatter plot suggests a positive relationship between experience and salary, although further analysis may be needed to confirm this trend.
